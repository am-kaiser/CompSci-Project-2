```python
import question3_project2_elem as q3p2
import os
import autograd.numpy.random as npr
import autograd.numpy as np
import time
import matplotlib.pyplot as plt
```

```python
# Checking if the parameters learnt which has the lowest error is able to learn unseen training data
Nx = 100
Nt = 100
x = np.linspace(0, 1, Nx)
t = np.linspace(0, 1, Nt)
xx, tt = np.meshgrid(x, t)
```

# Experiment 9

```python
#Using parameters of GRID 5*5 NonSGD, lambda = 0.1 with a single hidden layer with 100 neurons
#experiment number 9
output_dir = 'output/'
expID = 9
postfix = str(expID)

num_neurons = np.load(output_dir + 'num_neurons_' + postfix + '.npy')
N_hidden = np.size(num_neurons)
w_and_b = [None]*(N_hidden + 1)
for layer in range(len(num_neurons)+1):
    w_and_b[layer] = np.load(output_dir + 'w_and_b_' + postfix + '_layer_' + str(layer+1) + '.npy')

g_nn = np.zeros((Nt, Nx))
g_an = np.zeros((Nt, Nx)) #analytic solution
g9=g_nn
for ix, x_ in enumerate(x):
    for it, t_ in enumerate(t):
        g_nn[it, ix] = q3p2.g_trial(np.array([x_,t_]), w_and_b)
        g_an[it, ix] = q3p2.g_analytic(np.array([x_,t_]))
        
fig, ax = plt.subplots(1, 2, facecolor = 'w', figsize = (10, 4))
cont1 = ax[0].pcolormesh(tt, xx, g_an, cmap = 'hsv')#, vmax = max_value, vmin = min_value)
cont1 = ax[1].pcolormesh(tt, xx, g_nn, cmap = 'hsv')#, vmax = max_value, vmin = min_value)
fig.colorbar(cont1, ax = ax)
#plt.title('Number of space points: ' + str(Nx) + '\n' + 
#          'Number of time points: ' + str(Nt) + '\n')
ax[0].set_title('Analytical Solution')
ax[1].set_title('Neural Network Solution')
ax[0].set_ylabel('x')
ax[0].set_xlabel('t')
ax[1].set_ylabel('x')
ax[1].set_xlabel('t')

plt.savefig('unseen5y5.png', dpi=250,bbox_inches='tight')
"""
fig, ax = plt.subplots(facecolor = 'w')
cont = ax.pcolormesh(xx, tt, g_an, cmap = 'hsv')
fig.colorbar(cont, ax = ax)
plt.title('Number of space points: ' + str(Nx) + '\n' + 
          'Number of time points: ' + str(Nt) + '\n')

fig, ax = plt.subplots(facecolor = 'w')
cont = ax.pcolormesh(xx, tt, g_nn, cmap = 'hsv')
fig.colorbar(cont, ax = ax)
plt.title('Number of space points: ' + str(Nx) + '\n' + 
          'Number of time points: ' + str(Nt) + '\n')
"""
fig, ax = plt.subplots(facecolor = 'w', figsize = (5,4))
cont = ax.pcolormesh(tt, xx, g_an-g_nn, cmap = 'hsv')
fig.colorbar(cont, ax = ax)
plt.title('Error')
ax.set_ylabel('x')
ax.set_xlabel('t')
plt.savefig('unseenerror5by5.png', dpi=250,bbox_inches='tight')
print('Max abs error:', np.max(np.abs(g_an-g_nn)))
print('Mean abs error:', np.mean(np.abs(g_an-g_nn)))
print('Mean Squared error:', np.max(np.mean((g_an-g_nn)**2)))

np.save('NeuralNetwork_5x5.npy', g_nn)
```

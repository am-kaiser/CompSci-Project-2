```python
def plot_error_field(exp):
    g_an = np.load(output_dir + 'g_an_' + str(exp) + '.npy')
    g_nn = np.load(output_dir + 'g_nn_' + str(exp) + '.npy')
    error = g_an - g_nn
    Nx = experiments[exp][0]
    Nt = experiments[exp][1]
    n_iter = experiments[exp][2]
    lmb = experiments[exp][3]
    num_neurons = experiments[exp][4]
    SGD = experiments[exp][5]
    
    x = np.linspace(0, 1, Nx)
    t = np.linspace(0, 1, Nt)
    xx, tt = np.meshgrid(x, t)
    
    fig, ax = plt.subplots(facecolor = 'w')
    cont = ax.pcolormesh(xx, tt, error, cmap = 'Spectral',vmax=0.1, vmin = -0.3)
    ax.set_ylabel('x')
    ax.set_xlabel('t')
    fig.colorbar(cont, ax = ax)
    plt.title('Error')
def plot_absolute_error_field(exp):
    g_an = np.load(output_dir + 'g_an_' + str(exp) + '.npy')
    g_nn = np.load(output_dir + 'g_nn_' + str(exp) + '.npy')
    
    error = np.abs(g_an - g_nn)
    Nx = experiments[exp][0]
    Nt = experiments[exp][1]
    n_iter = experiments[exp][2]
    lmb = experiments[exp][3]
    num_neurons = experiments[exp][4]
    SGD = experiments[exp][5]
    
    x = np.linspace(0, 1, Nx)
    t = np.linspace(0, 1, Nt)
    xx, tt = np.meshgrid(x, t)
    
    fig, ax = plt.subplots(facecolor = 'w')
    cont = ax.pcolormesh(xx, tt, error, cmap = 'hsv')
    ax.set_ylabel('x')
    ax.set_xlabel('t')
    fig.colorbar(cont, ax = ax)
    plt.title('Error')
```

# Plotting Functions

```python
def plot_analytical(exp):
    g_an = np.load(output_dir + 'g_an_' + str(exp) + '.npy')
    Nx = experiments[exp][0]
    Nt = experiments[exp][1]
    n_iter = experiments[exp][2]
    lmb = experiments[exp][3]
    num_neurons = experiments[exp][4]
    SGD = experiments[exp][5]
    
    x = np.linspace(0, 1, Nx)
    t = np.linspace(0, 1, Nt)
    xx, tt = np.meshgrid(x, t)
    
    fig, ax = plt.subplots(facecolor = 'w')
    cont = ax.pcolormesh(xx, tt, g_an, cmap = 'hsv')
    ax.set_ylabel('x')
    ax.set_xlabel('t')
    fig.colorbar(cont, ax = ax)
    plt.title('Number of space points: ' + str(Nx) + '\n' + 
              'Number of time points: ' + str(Nt) + '\n')
def plot_neural(exp):
    g_nn = np.load(output_dir + 'g_nn_' + str(exp) + '.npy')
    Nx = experiments[exp][0]
    Nt = experiments[exp][1]
    n_iter = experiments[exp][2]
    lmb = experiments[exp][3]
    num_neurons = experiments[exp][4]
    SGD = experiments[exp][5]
    
    x = np.linspace(0, 1, Nx)
    t = np.linspace(0, 1, Nt)
    xx, tt = np.meshgrid(x, t)
    
    fig, ax = plt.subplots(facecolor = 'w')
    cont = ax.pcolormesh(xx, tt, g_nn, cmap = 'hsv')
    ax.set_ylabel('x')
    ax.set_xlabel('t')
    fig.colorbar(cont, ax = ax)
    plt.title('Number of space points: ' + str(Nx) + '\n' + 
              'Number of time points: ' + str(Nt) + '\n')
    
def plot_solution(exp):
    g_an = np.load(output_dir + 'g_an_' + str(exp) + '.npy')
    g_nn = np.load(output_dir + 'g_nn_' + str(exp) + '.npy')
    
    Nx = experiments[exp][0]
    Nt = experiments[exp][1]
    n_iter = experiments[exp][2]
    lmb = experiments[exp][3]
    num_neurons = experiments[exp][4]
    SGD = experiments[exp][5]
    
    x = np.linspace(0, 1, Nx)
    t = np.linspace(0, 1, Nt)
    xx, tt = np.meshgrid(x, t)
    
    max_value = np.max(np.array([np.max(g_an), np.max(g_nn)]))
    min_value = np.max(np.array([np.min(g_an), np.min(g_nn)]))
    
    fig, ax = plt.subplots(1, 2, facecolor = 'w', figsize = (20, 8))
    cont1 = ax[0].pcolormesh(xx, tt, g_an, cmap = 'hsv', vmax = max_value, vmin = min_value)
    cont1 = ax[1].pcolormesh(xx, tt, g_nn, cmap = 'hsv', vmax = max_value, vmin = min_value)
    fig.colorbar(cont1, ax = ax)
    #plt.title('Number of space points: ' + str(Nx) + '\n' + 
    #          'Number of time points: ' + str(Nt) + '\n')
    ax[0].set_title('Analytical Solution')
    ax[1].set_title('Neural Network Solution')
    ax[0].set_ylabel('x')
    ax[0].set_xlabel('t')
    ax[1].set_ylabel('x')
    ax[1].set_xlabel('t')
    #plt.rcParams.update({'font.size': 22})
```

```python
def plot_cost_convergence(expList):
    n_iter = 1000
    cost_arr = np.zeros([len(expList), n_iter])
    
    for ind_e, exp in enumerate(expList):
        cost_arr[ind_e] = np.load(output_dir + 'cost_arr_' + str(exp) + '.npy')
    
    fig, ax = plt.subplots(facecolor = 'w')
    for ind_e, exp in enumerate(expList):
        ax.plot(cost_arr[ind_e], label = exp)
        ax.set_ylim(0, 100)
    ax.legend()
```
